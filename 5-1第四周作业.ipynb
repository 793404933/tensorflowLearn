```python
#自己写入 加一层隐藏层
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
#载入数据集
mnist = input_data.read_data_sets("C:\\Users\\Administrator\\Desktop\\MNIST_data",one_hot=True)

batch_size = 100
n_batch = mnist.train.num_examples // batch_size
#定义占位符
x = tf.placeholder(tf.float32,[None,784])
y = tf.placeholder(tf.float32,[None,10])
#定义神经网络  （784，100） （100，10）
# 定义神经网络中间层
Weights_L1 = tf.Variable(tf.truncated_normal([784,200],stddev=0.1))
biases_L1 = tf.Variable(tf.zeros([1,200])+0.1)
Wx_plus_b_L1 = tf.matmul(x, Weights_L1) + biases_L1
#激活函数
L1 = tf.nn.relu(Wx_plus_b_L1)
#定义神经网络输出层
Weights_L2 = tf.Variable(tf.truncated_normal([200,10],stddev=0.1))
biases_L2 = tf.Variable(tf.zeros([1,10])+0.1)
L2 = tf.nn.relu(tf.matmul(L1,Weights_L2)+biases_L2)
prediction = tf.nn.softmax(L2)
#定义二次损失函数
# loss = tf.reduce_mean(tf.square(y-prediction))
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y,logits=L2))
# 使用梯度下降法
#train_step = tf.train.GradientDescentOptimizer(0.2).minimize(loss)
train_step = tf.train.MomentumOptimizer(0.2,momentum=0.5).minimize(loss)
#初始化变量
init = tf.global_variables_initializer()

# 定义求准确率的方法
# tf.argmax比较参数  ；tf.argmax求标签最大的值，在那个位置，相当于标签
# 结果存放在一个布尔型列表中
correct_prediction = tf.equal(tf.argmax(y,axis = 1),tf.argmax(prediction,axis = 1))# argmax返回最大值所在位置
# 求准确率
# 转换格式tf.cast;然后求平均值即可得到准确率；
accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))
with tf.Session() as sess:
    sess.run(init)
    for epoch in range(50):
        for batch in range(n_batch):
            #图片数据，图片标签；
            batch_xs,batch_ys = mnist.train.next_batch(batch_size)
            sess.run(train_step,feed_dict={x:batch_xs,y:batch_ys})
        acc = sess.run(accuracy,feed_dict={x:mnist.test.images,y:mnist.test.labels})
        print("Iter"+str(epoch)+",Testing Accuracy:"+str(acc))
```

```python
#自己写入 加一层隐藏层
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
#载入数据集
mnist = input_data.read_data_sets("C:\\Users\\Administrator\\Desktop\\MNIST_data",one_hot=True)

batch_size = 100
n_batch = mnist.train.num_examples // batch_size
#定义占位符
x = tf.placeholder(tf.float32,[None,784])
y = tf.placeholder(tf.float32,[None,10])
keep_prob = tf.placeholder(tf.float32)
lr = tf.Variable(0.01,dtype = tf.float32)

#定义神经网络  （784，100） （100，10）
# 定义神经网络中间层
Weights_L1 = tf.Variable(tf.truncated_normal([784,500],stddev=0.1))
biases_L1 = tf.Variable(tf.zeros([1,500])+0.1)
Wx_plus_b_L1 = tf.matmul(x, Weights_L1) + biases_L1
#激活函数
L1 = tf.nn.relu(Wx_plus_b_L1)
L1_drop = tf.nn.dropout(L1,keep_prob)

#定义神经网络输出层
Weights_L2 = tf.Variable(tf.truncated_normal([500,300],stddev=0.1))
biases_L2 = tf.Variable(tf.zeros([1,300])+0.1)
L2 = tf.nn.relu(tf.matmul(L1_drop,Weights_L2)+biases_L2)
L2_drop = tf.nn.dropout(L2,keep_prob)

Weights_L3 = tf.Variable(tf.truncated_normal([300,10],stddev=0.1))
biases_L3 = tf.Variable(tf.zeros([1,10])+0.1)
prediction = tf.nn.softmax(tf.matmul(L2_drop,Weights_L3)+biases_L3)
#定义二次损失函数
# loss = tf.reduce_mean(tf.square(y-prediction))
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y,logits=prediction))
# 使用梯度下降法
#train_step = tf.train.GradientDescentOptimizer(0.2).minimize(loss)
train_step = tf.train.AdamOptimizer(lr).minimize(loss)
#初始化变量
init = tf.global_variables_initializer()

# 定义求准确率的方法
# tf.argmax比较参数  ；tf.argmax求标签最大的值，在那个位置，相当于标签
# 结果存放在一个布尔型列表中
correct_prediction = tf.equal(tf.argmax(y,axis = 1),tf.argmax(prediction,axis = 1))# argmax返回最大值所在位置
# 求准确率
# 转换格式tf.cast;然后求平均值即可得到准确率；
accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))
with tf.Session() as sess:
    sess.run(init)
    for epoch in range(50):
        sess.run(tf.assign(lr,0.001*(0.95 ** epoch)))
        for batch in range(n_batch):
            #图片数据，图片标签；
            batch_xs,batch_ys = mnist.train.next_batch(batch_size)
            sess.run(train_step,feed_dict={x:batch_xs,y:batch_ys,keep_prob:1.0})
        learning_rate = sess.run(lr)
        acc = sess.run(accuracy,feed_dict={x:mnist.test.images,y:mnist.test.labels,keep_prob:1.0})
        print("Iter"+str(epoch)+",Testing Accuracy:"+str(acc)+",Learning rate:"+str(learning_rate))
```
